```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from tabulate import tabulate
%matplotlib inline
```

# Read and analyze data

Features/label description:
* _season_: 1 - spring, 2 - summer, 3 - autumn, 4 - winter
* _yr_: 0 - 2011, 1 - 2012
* _mnth_: 1 to 12
* _holiday_: 0 - no holiday, 1 - there is a holiday
* _weekday_: 0 to 6
* _workingday_: 0 - non-working day, 1 - working day
* _weathersit_: weather rating from 1 (clean, clear day) to 4 (rain, fog)
* _temp_: temperature in Celsius
* _atemp_: temperature sensations in Celsius
* _hum_: humidity
* _windspeed (mph) _: wind speed in miles per hour
* _windspeed (ms) _: wind speed in meters per second
* _cnt_: the number of rented bicycles (label)

```python
path = os.path.join(
    'datasets',
    'bikes_rent.csv'
)
df = pd.read_csv(path, sep = ',')
df.head()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(15, 10))
for idx, feature in enumerate(df.columns[:-1]):
    index = int(idx / 4)
    df.plot(feature, "cnt", subplots=True, kind="scatter", ax=axes[index, idx % 4])
```

```python
# Correlations with target label
features_columns = df.columns[:-1]
features = df[features_columns]
features.corrwith(df['cnt'])
```

```python
# Correlations between similar factors
columns = ['temp', 'atemp', 'hum', 'windspeed(mph)', 'windspeed(ms)', 'cnt']
df[columns].corr()
```

```python
# Features means
features.mean()
```

```python
from sklearn.preprocessing import scale
from sklearn.utils import shuffle
```

```python
# Shuffle data and scale
df_shuffled = shuffle(df, random_state=123)
X = scale(df_shuffled[df_shuffled.columns[:-1]])
y = df_shuffled["cnt"]
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
def print_features_weights(features_names, weights):
    """Print features names and corresponded weights.
    
    Args:
        features_names (list): list of features names.
        weights (numpy.ndarray): corresonded weights.
        
    """
    
    print(
        tabulate(
            zip(features_names, weights),
            headers = ['feature', 'coef'],
            tablefmt='grid',
            stralign='center'
        )
    )
```

```python
regr = LinearRegression()
regr.fit(X, y)
print_features_weights(list(df_shuffled.columns[:-1]), regr.coef_)
```

Linear dependency weights are too big...

Regularization types:

$L1 \Longleftrightarrow Lasso$

$L2 \Longleftrightarrow Ridge$

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
lasso_regressor = Lasso()
lasso_regressor.fit(X, y)
print_features_weights(list(df_shuffled.columns[:-1]), lasso_regressor.coef_)
```

```python
ridge_regressor = Ridge()
ridge_regressor.fit(X, y)
print_features_weights(list(df_shuffled.columns[:-1]), ridge_regressor.coef_)
```

# Visualise dependencies between alphas and coefficients

```python
alphas = np.arange(1, 500, 50)
coefs_lasso = np.zeros((alphas.shape[0], X.shape[1]))
coefs_ridge = np.zeros((alphas.shape[0], X.shape[1]))

for count, alpha in enumerate(alphas):
    lasso_regressor = Lasso(alpha=alpha)
    lasso_regressor.fit(X, y)
    coefs_lasso[count, :] = lasso_regressor.coef_
    
    ridge_regressor = Ridge(alpha=alpha)
    ridge_regressor.fit(X, y)
    coefs_ridge[count, :] = ridge_regressor.coef_
```

```python
plt.figure(figsize=(8, 5))
for coef, feature in zip(coefs_lasso.T, df.columns):
    plt.plot(alphas, coef, label=feature, color=np.random.rand(3))
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("alpha")
plt.ylabel("feature weight")
plt.title("Lasso")

plt.figure(figsize=(8, 5))
for coef, feature in zip(coefs_ridge.T, df.columns):
    plt.plot(alphas, coef, label=feature, color=np.random.rand(3))
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("alpha")
plt.ylabel("feature weight")
plt.title("Ridge")
```

# LassoCV

```python
from sklearn.linear_model import LassoCV
```

```python
alphas = np.arange(1, 100, 5)
lasso_cv_regressor = LassoCV(alphas = alphas, cv = 3)
lasso_cv_regressor.fit(X, y)
print("optimum alpha: ", lasso_cv_regressor.alpha_)
print_features_weights(list(df_shuffled.columns[:-1]), lasso_cv_regressor.coef_)
```

Some interpretations:
1. The last trained model: lasso_cv_regressor. 4 features with the largest positive coefficients:
    
    feature: yr, weight: 1015.06
    
    feature: atemp, weight: 581.397
    
    feature: season, weight: 532.019
    
    feature: temp, weight: 370.68
    
    On the average, dependency is visible. But only in the case of temperatures we can speak about a linear dependence. From common sense, the year should not affect linearly the bikes rent quantity (we need more information about these years). The time of the year, indeed, should influence, but here, the dependence may not be linear, but seasonal (for example, in the warm season, bikes rent may be more popular). The temperature may influence linearly. With increasing temperature, people may buy bicycles more often.
    
    
2. 3 characteristics with the most negative coefficients in absolute value:
   
   feature: weathersit, weight: -330.56 - it makes sense. With the higher values of this factor the weather is worse by definition. Therefore it is less likely that people will want to ride a bike
    
    feature: windspeed(mph), weight: -191.771 - it makes sense too. With a high wind speed, it's much harder to ride. However, on the graph this is rather a cloud of points, and it will not be possible to see a linear dependence on it.
    
    feature: hum, weight: -140.007 - it makes sense too. At high humidity, heat loss deteriorates, making people feel hot and uncomfortable driving. However, again, on the graph this is a cloud of points, the dependence is hard to read from it.
    
    feature: mnth, weight: -100.04 - here, as in the case of season, the dependence should rather be nonlinear or seasonal, which can be read from the graph.
    
    
3. Windspeed(ms) has a much less value than other factors in absolute terms. This may be due to the presence of windspeed (mph), which are linearly related. The model traced the multicollinearity, and the coefficient was reset.
